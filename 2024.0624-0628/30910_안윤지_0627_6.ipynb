```python
# 꺽은선 그래프 그리기 
# 라이브러리 불러오기 및 별명 지정하기

import matplotlib.pyplot as plt

# 한글폰트 적용하기
plt.rc('font' ,family = 'Malgun gothic')

#x,y축 데이터 지정
sales = [1,5,7,3,7]
month = ['mar','apr','may','jun','jul']

#그래프의 해상도 설정하기 dpi.figure
plt.figure(dpi = 100)

# 제목 title
plt.title('월별 판매 실적')

#꺽은선형 그래프 그리기 .plot.color
# 그래프 범례 .label
plt.plot(month,sales,color = 'm',label = '단위:개수')

# 범례보이게 설정 위치 .lower right
plt.legend(loc = 'lower right')
# x축 , y축 xlabel, ylabel
plt.xlabel('월')
plt.ylabel('매출')

# 그래프 보이게
plt.show()
```

```python
# 라이브러리 불러오기 및 별명 지정하기
import matplotlib.pyplot as plt
plt.rc('font', family = 'Malgun Gothic')

# data
sales = [1,5,7,3,7]
month = [3,4,5,6,7]

# 막대 그래프 그리기 plt.bor()
plt.title('월별 판매 실적') # 제목
plt.bar(month,sales,color = 'b') # 컬러 파랑

# x축 , y축 xlabel,ylabel
plt.xlabel('월')
plt.ylabel('매출')

plt.show()
```

```python
import random
import matplotlib.pyplot as plt
# 한글폰트 적용하기
plt.rc('font', family = 'Malgun Gothic')

a = []
for i in range(10):
    a.append(random.randint(1,10))
print(a)

# 제목
plt.title('히스토그램 차트 그리기')

# 히스토그램 차트 그리기
plt.hist(a)
plt.show()
```

```python
# 라이블러리 불러오기 및 별명 지정하기
import matplotlib.pyplot as plt
# 한글 폰트 적용하기
plt.rc('font', family = 'Malgun Gothic')

# data 혈액형 수
b_type = [25,19,37,11]
b_name = ['A','b','O','AB']

# 제목 : 혈액형 분포
plt.title('혈액형 분포')

# 원형 차트 그리는 명령어 ple 범례 labels = [] autopct = ''
plt.pie(b_type, labels = b_name, autopct = '%1.1f%%')
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.rc('font', family = 'Malgun Gothic')

import random

height = []
weight = []
for i in range(100):
    height.append(random.randint(100,200))
    weight.append(random.randint(20,100))
    
plt.title('키와 몸무게 상관관계')

plt.scatter(height,weight, c = height,cmap = 'RdPu')

plt.colorbar(label = '키')
plt.xlabel('키')
plt.ylabel('몸무게')

plt.show()
```

```python
import matplotlib.pyplot as plt

plt.rc('font', family = 'Malgun Gothic')

import random

height,weight,score = [],[],[]
for i in range(100):
    height.append(random.randint(100,200))
    weight.append(random.randint(20,100))
    score.append(random.randint(1,1000))
    
plt.title('키와 몸무게의 상관관계')

plt.scatter(height,weight, s = score, c = score, cmap ='rainbow', alpha = 0.7 )

plt.colorbar(label = '체육 실기 점수')
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.rc('font', family = 'Malgun Gothic')

singer = ['A','B','C','D','E']
week1 = [42,58,19,92,84]
week2 = [53,52,48,98,73]

for i in range(len(week1)):
    week1[i] = week1[i]

plt.rcParams['axes.unicode_minus'] = False

plt.title('오디션 프로그램 득표 현황')

plt.barh(singer,week1,label = '첫째주', color = 'hotpink')
plt.barh(singer,week2,label = '둘째주', color = 'royalblue')

plt.legend()
plt.show()
```
