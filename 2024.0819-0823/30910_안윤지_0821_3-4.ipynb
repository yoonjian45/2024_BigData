# 함수와 메이드 차이 알아보기

```
- 내장함수 : 프로그램안에 마리 정의되어 있는 함수
    sum,max,print,min
- 패키지 함수 : 별도로 설치, 불러와야 하는 함수
    import pandas as pd
    import matplotlib.pyplot as plt
- 메서드 : 변수가 지니고 있는 함수 이름과 괄호 같이 사용
    df.head() .info() .tail() .shape() .decribe()
    메서드  = 기술 , 어트리뷰트 = 능력치
```
