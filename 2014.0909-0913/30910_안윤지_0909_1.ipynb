```python
mpg = pd.read_csv('mpg.csv')
mpg.head()
```

```python
#   
mpg['total'].plot.hist()
```

```python
# 20  pass,   fail  np.where = if

mpg['test'] = np.where(mpg['total'] >= 20, 'pass', 'fail')
mpg.head()
```

```python
#    

mpg['test'].value_counts()
```

```python
count_test = mpg['test'].value_counts()
count_test.plot.bar(rot = 0)
```

```python
#  total  A,B,C A 30 B 20  C
# colum name grade

mpg['grade'] = np.where(mpg['total'] >= 30, 'A',
                       np.where(mpg['total'] >= 20, 'B','C'))
mpg.head()
```

```python
count_grade = mpg['grade'].value_counts().sort_index()
count_grade
```

```python
count_grade.plot.bar(rot = 0)
```

```python
# A,B,C,D  
# grade2  where 30 A 25 B 20 C  D

mpg['grade2'] = np.where(mpg['total'] >= 30, 'A',
                        np.where(mpg['total'] >= 25 ,'B',
                                np.where(mpg['total'] >= 20,'C','D')))
```

```python
count_grade2 = mpg['grade2'].value_counts().sort_index()
count_grade2
```

```python
count_grade2.plot.bar(rot = 0)
```

```python
# suv  query
#     .assign total   (hwy+cty)/2
#    .groupby
#     .agg mean_tot
#   .sort_values(ascending= False)
# 1~5  .head()

mpg.query('category == "suv"')\
.assign(total = mpg['hwy'] + mpg['cty']/2)\
.groupby('manufacturer')\
.agg(mean_tot = ('total','mean'))\
.sort_values('mean_tot',ascending = False)\
.head()
```
