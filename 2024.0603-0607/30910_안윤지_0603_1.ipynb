```python
# 꺽은선 그래프 그리기
# 라이브러리 불러오기 및 별명 지정하기

import matplotlib.pyplot as plt
# 한글 폰트 적용하기
plt.rc('font', family = 'Malgun Gothic')

# x,y축 데이터 지정
sales = [1,5,7,3,7]
month = ['mar','apr','may','jun','jul']

# 그래프의 해상도 설정하기 dpi figure
plt.figure(dpi = 100)

# 제목 title 
plt.title('월별 판매 실적')

# 꺽은 선형 그래프 그리기.plot.color
# 그래프 범래.label
plt.plot(month,sales, color = 'm', label = '단위:개수')

# 범레 보이게 설정 위치 .lower right
plt.legend(loc = 'lower right')
# x축,y축 xlaber,ylaber
plt.xlaber()
plt.ylaber()
# 그래프 보이기
plt.show()
```
