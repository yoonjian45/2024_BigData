## 원형차트 그리기

```python
import matplotlib.pyplot as plt
plt.rc('font', family = 'Malgun Gothic')

# data 혈액형 수 
b_type = [25,19,37,11]

# 원형차트 그리기 명령어 plt
plt.pie(b_type)
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.rc('font', family = 'Malgun Gothic')

# data 혈액형 수 
b_type = [25,19,37,11]
b_name = ['A형','B형','O형','AB형']

# 원형차트 그리기 명령어 plt 범레 labels = []
plt.pie(b_type, labels = b_name)
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.rc('font', family = 'Malgun Gothic')

# data 혈액형 수 
b_type = [25,19,37,11]
b_name = ['A형','B형','O형','AB형']

# 원형차트 그리기 명령어 plt 범레 labels = [] autopct = ''
plt.pie(b_type, labels = b_name, autopct = '%1.1f%%')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.rc('font', family = 'Malgun Gothic')

# data 혈액형 수 
b_type = [25,19,37,11]
b_name = ['A형','B형','O형','AB형']

# 제목 : 형액형 분포
plt.title('혈액형 분포')

# 원형차트 그리기 명령어 plt 범레 labels = [] autopct = ''
plt.pie(b_type, labels = b_name, autopct = '%1.1f%%')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.rc('font', family = 'Malgun Gothic')

# data 국어 영어
kor = [80,20,50,20,10,50,60,30,40]
eng = [90,40,10,40,60,20,50,70,90]

# 산점도 그리기 scatter option color = 'r' 투명도 alpha 값
plt.scatter(kor,eng , color = 'r',alpha = 0.5)
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.rc('font', family = 'Malgun Gothic')

# data 국어 영어
kor = [80,20,50,20,10,50,60,30,40]
eng = [90,40,10,40,60,20,50,70,90]

# 산점도 그리기 scatter option color = 'r' 투명도 alpha 값
plt.scatter(kor,eng , color = 'r',alpha = 0.5)

# 축제목 x,y축
plt.xlabel('국어점수')
plt.ylabel('영어점수')

plt.show()
```
