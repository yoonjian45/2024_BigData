```python
import numpy as np
import pandas as pd
```

```python
exam = pd.read_csv('exam.csv')
exam.head()
```

```python
mpg = pd.read_csv('mpg.csv')
mpg.head()
```

```python
# 집단별 요약하기
# exam math평균 구하기

exam.agg(mean_math = ('math','mean'))
```

### 집단별로 요약 동계량 구하기

```python
# nclass 별로 분리하기 .gorupby()
# math 평균 .agg
# 반별 수학 평균
exam.groupby('nclass').agg(mean_math = ('math','mean'))
```

```python
# 반별 영어 평균
exam.groupby('nclass').agg(mean_english = ('english','mean'))
```

```python
# 반별 과목 평균
exam.groupby('nclass')\
.agg(mean_english = ('english','mean'),
    mean_math = ('math','mean'),
    science = ('science','mean'))
```

```python
# nclass 별로 분리
# 수학 점수 평균
# 수학 점수 합계
# 수학 점수 중앙값
# 빈도(학생 수)
exam.groupby('nclass')\
.agg(mean_math = ('math','mean'),
    sum_math = ('math','sum'),
    median_math = ('math','median'),
    n = ('nclass','count'))
```

```python
# 모든 변수의 요약 통계량 한번에 구하기
# 과목별 평균 
exam.groupby('nclass').mean()
```

```python
# 제조회사, 구동방식별로 구분

mpg.groupby(['manufacturer','drv'])\
.agg(mean_cty = ('cty','mean'))
```

```python
# audi의 drv별 빈도
# audi추출
# drv 별 분리
# 빈도
mpg.query('manufacturer == "audi"')\
.groupby(['drv'])\
.agg(n = ('drv','count'))
```

```python
## chevrolet의 drv별 빈도

# chevrolet 추출
# drv별 분리
# 빈도 구하기
mpg.query('manufacturer == "chevrolet"') \
       .groupby(['drv']) \
       .agg(n = ('drv', 'count'))
```

```python
# drv별 분리
# 빈도 구하기

mpg.groupby('drv')\
.agg(n = ('drv','count'))
```

```python
# value_count
mpg['drv'].value_counts()
```

```python
# drv 빈도 구하기
# 데이터 프레임으로 바꾸기, 변수명 n으로 바꾸기 to_frame
# axis 이름에 drv 지정
# n이 100을 초과한 경우 추출

mpg['drv'].value_counts()\
.to_frame('n')\
.query('n > 100')
```

```python
# suv 추출 query
# 합산 연비 변수 만들기 .assign
# 재조 회사별로 분리 .groupby
# 함산 연비 평균 구하기 .agg mean_tot
# 내림차순 정렬 .sort_values(ascending = False)
# 1~5위 까지 출력 .head()

mpg.query('category == "suv"')\
.assign(total = (mpg['hwy']+mpg['cty'])/2)\
.groupby('manufacturer')\
.agg(mena_tot = ('total','mean'))\
.sort_values('mena_tot',ascending = False)\
.head()
```
