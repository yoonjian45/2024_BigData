```python
import pandas as pd
import numpy as np
```

```python
exam = pd.read_csv('exam.csv')
exam.head()
```

```python
mpg = pd.read_csv('mpg.csv')
mpg.head()
```

```python
# total 변수 추가 .assign
exam.assign(total = exam['math']+exam['english']+exam['science']).head()
```

```python
# mean 변수 추가
exam.assign(total = exam['math']+exam['english']+exam['science'],
           mean = (exam['math']+exam['english']+exam['science'])/3).head()
```

```python
# np.where 과학 점수가 60점 이상이면 pass fall출력
# teat
exam.assign(test = np.where(exam['science']>=60,'pass','fall')).head()
```

```python
# 긴 데이터프레임명 지정

long_name = pd.read_csv('exam.csv')
long_name.head()
```

```python
# long_name new변수 수학 영어 과학 점수 합계
long_name.assign(new = long_name['math'] + long_name['english'] + long_name['science']).head()
```

```python
# long_name 대신 x 입력
long_name.assign(new = lambda x:x['math'] + x['english'] + x['science']).head()
```

```python
# mean 변수 추가
exam.assign(total = exam['math']+exam['english']+exam['science'],
           mean = lambda x:x['total']/3).head()
```

```python
exam.assign(total = lambda x:x['math']+x['english']+x['science'],
            mean = lambda x:x['total']/3).head().head()
```

```python
# 요약하기 agg
# math 평균구하기
exam.agg(mean_math = ('math','mean'))
```

```python
# class별 요약 통계량
# nclass별
# math 평균

exam.groupby('nclass') \
    .agg(mean_math = ('math','mean'))
```

```python
exam.groupby('nclass',as_index = False) \
    .agg(mean_math = ('math','mean'))
```

```python
exam.groupby('nclass').mean()
```
